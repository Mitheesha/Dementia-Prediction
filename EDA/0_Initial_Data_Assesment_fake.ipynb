#### 3.2 Isolating Medical columns

##### 3.2.1 Getting values of ranges of different sections of medical variables

```python
# Defining medical coloumn ranges
medical_ranges = {
    'A4_Subject_Medications': ('ANYMEDS', 'DRUG40'), # A4
    'B_Section_Physical_to_ClinicianSymptoms': ('HEIGHT', 'FTLDEVAL'), # B section
    'C_Section_Neuropsych_Summary': ('MMSECOMP', 'RESPOTHX'), # C section
    'D_Section_Diagnosis_MedicalConditions': ('WHODIDDX', 'OTHCONDX') # D section
}
```

```python
cols = list(df.columns)

for name, (start, end) in medical_ranges.items():
    print(f"{name}")
    print(f"Start: {start} | End: {end}")

    if start not in cols or end not in cols:
        print(f"One of the boundaries not found in dataframe: {start}, {end}")
        continue

    start_idx = cols.index(start)
    end_idx = cols.index(end)

    selected_cols = cols[start_idx : end_idx + 1]

    print(f"Total columns: {len(selected_cols)}")
    print(selected_cols)
    print("\n")
```

##### 3.2.2 Identify duplicate variables comparing each range

```python
range_cols = {}

for name, (start, end) in medical_ranges.items():
    if start in cols and end in cols:
        start_idx = cols.index(start)
        end_idx = cols.index(end)
        selected_cols = cols[start_idx : end_idx + 1]
        range_cols[name] = set(selected_cols)
    else:
        range_cols[name] = set()  # empty if boundaries not found

# Check for columns that occur in more than one range
for r1 in range_cols:
    for r2 in range_cols:
        if r1 < r2:  # avoid self-comparison and duplicates
            common = range_cols[r1].intersection(range_cols[r2])
            print(f"Common columns between {r1} and {r2}: {len(common)}")
            if common:
                print(common, "\n")
```
